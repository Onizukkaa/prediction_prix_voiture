```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv("train.csv")
train
```

```python
train.drop(columns="New_Price",inplace=True)
```

```python
train.dropna(axis=0, inplace=True)
```

```python
train
```

```python
brand_col = train.Name.str.extract(pat=r'(\w+) (\w+)', expand=False)[0]
```

```python
brand_col
```

```python
brand_col.value_counts()
```

```python
train["Brand"] = brand_col
train
```

```python
brands_less_than_10 = train.Brand.value_counts()[train.Brand.value_counts().values < 10].index.to_list()
train.Brand = train.Brand.apply(lambda x: 'other' if x in brands_less_than_10 else x)
train
```

```python
train.info()
```

```python
train.Brand.value_counts()
```

```python
plt.plot(train.Year, train.Price)
plt.ylabel('Price')
plt.show()
```

```python
#plt.plot(train["Year"].loc[train["Brand"] == "Honda"], train["Price"].loc[train["Brand"] == "Honda"])
plt.figure(figsize=(8, 4))
plt.title('Manual prices')
plt.ylabel('Price')
sns.scatterplot(x=train["Year"].loc[train["Transmission"] == "Manual"], y=train["Price"].loc[train["Transmission"] == "Manual"], data=train)
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.title('Automatic prices')
plt.ylabel('Price')
sns.scatterplot(x=train["Year"].loc[train["Transmission"] == "Automatic"], y=train["Price"].loc[train["Transmission"] == "Automatic"], data=train)
plt.show()
```

```python
automatic_mean = train["Price"].loc[train["Transmission"] == "Automatic"].mean()
automatic_mean
```

```python
manual_mean = train["Price"].loc[train["Transmission"] == "Manual"].mean()
manual_mean
```

```python
plt.figure(figsize=(8, 4))
plt.title('Transmission mean prices')
plt.bar(["Automatic", "Manual"], [automatic_mean, manual_mean])
plt.show()
```

```python
engine_values = train.Engine.drop_duplicates()
engine_values
```

```python
plt.figure(figsize=(16, 4))
plt.title(str("Engine prices"))
plt.ylabel('Price')
sns.scatterplot(x = train["Engine"], y = train["Price"], data=train)
plt.show()
```

```python
for brand in brands:
    print(brand, " mean price : ", train["Price"].loc[train["Brand"] == brand].mean())
```

```python
for brand in brands:
    plt.figure(figsize=(8, 4))
    plt.title(str(brand))
    plt.ylabel('Price')
    sns.scatterplot(x = train["Year"].loc[train["Brand"] == str(brand)], y = train["Price"].loc[train["Brand"] == str(brand)], data=train)
    plt.show()
```
